##Missing values & Remove Unnecessary Columns:



```python
df["color"].fillna(df["color"].mode()[0], inplace=True)
df["director_name"].fillna("UnKnown", inplace=True)
df.dropna(subset=['num_critic_for_reviews'], inplace=True)
df["duration"].fillna(df["duration"].mean(), inplace=True)
df["director_facebook_likes"].fillna(df["director_facebook_likes"].mean(), inplace=True)
df["actor_3_facebook_likes"].fillna(df["actor_3_facebook_likes"].mean(), inplace=True)
df["actor_2_name"].fillna("UnKnown", inplace=True)
df["actor_1_facebook_likes"].fillna(df["actor_1_facebook_likes"].mean(), inplace=True)
df["actor_1_name"].fillna("UnKnown", inplace=True)
df["actor_3_name"].fillna("UnKnown", inplace=True)
df["facenumber_in_poster"].fillna(df["facenumber_in_poster"].mean(), inplace=True)
df["language"].fillna(df["language"].mode()[0], inplace=True)
df["country"].fillna(df["country"].mode()[0], inplace=True)
df["actor_2_facebook_likes"].fillna(df["actor_2_facebook_likes"].mean(), inplace=True)
df["gross"].fillna(0, inplace=True)
df["budget"].fillna(0, inplace=True)
df["title_year"].fillna(0, inplace=True)
df.drop(columns=["plot_keywords", "movie_imdb_link", "content_rating", "aspect_ratio"], axis=1, inplace=True)
df.isnull().sum()
```

## Analyze the distribution of movie genres and their impact on the IMDB score.

```python
df["genres"].value_counts(ascending=True)
```

```python
lis = ["Drama", "Comedy", "Romance", "Adventure", "Mystery", "Sci-Fi", "Thriller", "Animation", "Fantasy", "Crime", "Horror", "Sport", "Biography", "History",
       "Western", "Family", "Musical"]
mean_list = []
genres_name = []
counts = []
for i in lis:
  condition = df["genres"].str.contains(i, case=False, na=False)
  result_df = df.loc[condition, ["genres", "imdb_score"]]
  mean_list.append(result_df.imdb_score.mean())
  genres_name.append(i)
  counts.append(result_df.imdb_score.count())
key = {}
key["genres_name"] = genres_name
key["mean_IMBD"] = mean_list
key["counts"] = counts
print(key)
```

```python
graph = pd.DataFrame(key)
graph = graph.sort_values(by='mean_IMBD', ascending=False)
```

```python
import seaborn as sns

plt.figure(figsize=(12, 6))
ax = sns.barplot(x='genres_name', y='mean_IMBD', data=graph, palette='viridis')
for p, count in zip(ax.patches, graph['counts']):
    ax.annotate(f'Count:{count}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=7, color='black')
plt.title('Mean IMDB Ratings by Genre')
plt.xlabel('Genres')
plt.ylabel('Mean IMDB Ratings')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

##  Analyze the distribution of movie durations and its impact on the IMDB score.


```python
duration_df = df[df["duration"] != 0]
duration_df.shape
```

```python
Q1 = duration_df["duration"].quantile(0.25)
Q3 = duration_df["duration"].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for non-outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Use the clip function to cap values outside the bounds
duration_df["duration"] = duration_df["duration"].clip(lower=lower_bound, upper=upper_bound)
```

```python
a = duration_df[["duration", "imdb_score"]]
a.corr()
```

```python
plt.figure(figsize=(15, 8))
plt.scatter(duration_df['duration'], duration_df["imdb_score"])
plt.title('Movie Duration vs IMDB Ratings')
plt.xlabel('Movie Duration in (min)')
plt.ylabel('IMDB Ratings')
```

```python
duration_df['duration_interval'] = (duration_df['duration'] // 10) * 10
# Group by 'duration_interval' and calculate the mean IMDb rating
mean_ratings = duration_df.groupby('duration_interval')['imdb_score'].mean()

# Plotting the bar chart
plt.bar(mean_ratings.index, mean_ratings.values)

# Adding title and labels to the plot
plt.title('Mean IMDb Rating for Every 10-Minute Interval Of Movie Duration')
plt.xlabel('Duration Interval (minutes)')
plt.ylabel('Mean IMDb Rating')

# Display the plot
plt.show()
```
